```python
#finding IQR

percentile25=dataset['Final_weight'].quantile(0.35)
percentile75=dataset['Final_weight'].quantile(0.65)
print('25%:',percentile25)
print('75%:',percentile75)
iqr=percentile75-percentile25
print('IQR:',iqr)
```

```python
upper_limit = percentile75 + 1.5*iqr
lower_limit = percentile25 - 1.5*iqr

print('Upper Limit:',upper_limit)
print('Lower Limit:',lower_limit)
```

```python
dataset.describe()
```

```python
dataset[dataset['Final_weight']>upper_limit]
```

```python
dataset[dataset['Final_weight']<lower_limit]
```

```python
df = dataset[(dataset['Final_weight'] < upper_limit) & (dataset['Final_weight'] > lower_limit)]
```
